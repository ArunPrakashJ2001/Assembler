```python
def converter(a):
    val=bin(a)
    l=list(val)
    b=0
    fa=[]
    for i in l:
        if (i!='b'):
            b=b+1
        else:
            break
    l=l[b+1:]
    for j in range(16-len(l)):
        fa.append('0')
    val=fa+l
    return "".join(val)
```
